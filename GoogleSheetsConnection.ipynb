# Add markers for receivers to folium map

```python
import folium, os
from folium.features import CustomIcon


def add_receiver_marker(row):
    '''
    this function is setup to ingest a Pandas dataframe with Receiver, Latitude and Longitude columns
    and then create folium Marker objects to return
    '''
    url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/c/c1/{}'.format
    icon_image = url('Headphone_icon.svg/43px-Headphone_icon.svg.png')
    
    icon = CustomIcon(
        icon_image,
        icon_size=(30,30),
        icon_anchor=(10,10),
        popup_anchor=(-3, -76)
    )

    marker = folium.Marker(
        location=[row['Latitude'],row['Longitude']],
        icon=icon,
        popup=row['Receiver']
    )
    return marker
```

```python
shark.head()
```

```python
shark[['Receiver','Latitude','Longitude']]
```

```python
shark['receiver_markers'] = shark.apply(add_receiver_marker,axis=1)
```

```python
m = folium.Map(
    location=[27.4, -82.62],
    zoom_start=6,
    tiles='Stamen Terrain')
for marker in shark['receiver_markers']:
    m.add_child(marker)
    
folium.raster_layers.ImageOverlay(
    image='https://optics.marine.usf.edu/subscription/modis/GCOOS/2017/comp/248/A20172422017248.1KM.GCOOS.7DAY.L3D.CHL.png',
    name='7 Day Clorophyll, 8/30/2017 - 9/5/2017',
    bounds=[[18, -98.0], [31, -79]],
    opacity=0.5,
    interactive=False,
    cross_origin=False,
    zindex=1,
    alt='Wikipedia File:Mercator projection SW.jpg'
).add_to(m)

folium.LayerControl().add_to(m)
    
    
    
m
```

# PY Leaflet example 

```python
from pathlib import Path
path = Path('data')

from ipyleaflet import Map, Velocity, TileLayer, basemaps
import xarray as xr

zoom = 3

m = Map(
    center=[27.4, -82.62],
    zoom=6,
    interpolation='nearest',
    basemap=basemaps.CartoDB.DarkMatter,
)
```

```python
ds = xr.open_dataset(path.joinpath('wind-global.nc'))

display_options = {
    'velocityType': 'Global Wind',
    'displayPosition': 'bottomleft',
    'displayEmptyString': 'No wind data'
}
wind = Velocity(
    data=ds, 
    zonal_speed='u_wind', meridional_speed='v_wind', 
    latitude_dimension='lat', longitude_dimension='lon', 
    velocity_scale=0.01, max_velocity=20, 
    display_options=display_options
)
m.add_layer(wind)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap


def make_basemap(projection='robin', figsize=(9, 9), resolution='c'):
    fig, ax = plt.subplots(figsize=figsize)
    m = Basemap(projection=projection, resolution=resolution,
                lon_0=0, ax=ax)
    m.drawcoastlines()
    m.fillcontinents(color='0.85')
    return fig, m
```

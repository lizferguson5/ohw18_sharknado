```python
# use creds to create a client to interact with the Google Drive API
scope = ['https://spreadsheets.google.com/feeds']
creds = ServiceAccountCredentials.from_json_keyfile_name('../creds/oceanhackweek-cd26f3a7eca7.json', scope)
client = gspread.authorize(creds)

# Find a workbook by name and open the first sheet
# Make sure you use the right name here.
sheet = client.open_by_url("https://docs.google.com/spreadsheets/d/1MQsanWmsnlLaVUuaqaSObkHRety9tI2mrBQRLxHT80s/edit?usp=sharing").sheet1

list_of_hashes = sheet.get_all_records()
print(list_of_hashes)
```

```python
test = pd.DataFrame.from_records(list_of_hashes)
```

```python
test['Date and Time (UTC)'] = pd.to_datetime(test['Date and Time (UTC)'])
test.set_index('Date and Time (UTC)')
```

```python
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, output_file, reset_output, show, ColumnDataSource, save
from bokeh.models import BoxAnnotation
from bokeh.io import output_notebook 
```

```python
source = ColumnDataSource(test)
```

```python
# Set up plot
plot = figure(plot_height=400, plot_width=400, title="SharkNado!",
              tools="crosshair,pan,reset,save,wheel_zoom")

plot.line('Longitude', 'Latitude', source=source, line_width=3, line_alpha=0.6)
```

```python
from pathlib import Path
path = Path('data')

from ipyleaflet import Map, Velocity, TileLayer, basemaps
import xarray as xr

zoom = 3

m = Map(
    center=[27.4, -82.62],
    zoom=6,
    interpolation='nearest',
    basemap=basemaps.CartoDB.DarkMatter,
)
```

```python
ds = xr.open_dataset(path.joinpath('wind-global.nc'))

display_options = {
    'velocityType': 'Global Wind',
    'displayPosition': 'bottomleft',
    'displayEmptyString': 'No wind data'
}
wind = Velocity(
    data=ds, 
    zonal_speed='u_wind', meridional_speed='v_wind', 
    latitude_dimension='lat', longitude_dimension='lon', 
    velocity_scale=0.01, max_velocity=20, 
    display_options=display_options
)
m.add_layer(wind)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap


def make_basemap(projection='robin', figsize=(9, 9), resolution='c'):
    fig, ax = plt.subplots(figsize=figsize)
    m = Basemap(projection=projection, resolution=resolution,
                lon_0=0, ax=ax)
    m.drawcoastlines()
    m.fillcontinents(color='0.85')
    return fig, m
```
